```python
#import the necessary libraries
import cv2
from matplotlib import pyplot as plt
#Open an image
img = cv2.imread("../img/yieldsign.jpg")

#Convert image into RGB and Grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Create the environment for the picture
#output image to screen.
plt.subplot(1,1,1)
plt.imshow(img_rgb)
plt.show()
```

```python
stop_data = cv2.CascadeClassifier('stop_data.xml')
#Use detectMultiScale() function for OpenCV
found = stop_data.detectMultiScale(img_gray, minSize = (20, 20))

#Don't change anything if there are no signs.
amount_found = len(found)

if amount_found != 0:
    #Check if there is more than one sign in the image
    for (x,y,width,height) in found:
        #Draw a rectangle around every recognized sign
        cv2.rectangle(img_rgb, (x,y), (x + height, y + width), (0, 255, 0), 5)
#Create the environment of the picture and display it.
plt.subplot(1,1,1)
plt.imshow(img_rgb)
plt.show()
```
